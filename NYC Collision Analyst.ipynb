```python
import statistics
```

```python
numeric_df = df[['PEDESTRIAN_AFFECTED','CYCLIST_AFFECTED','MOTORIST_AFFECTED']]

for var in numeric_df:
    mean = statistics.mean(numeric_df[var])
    std = numeric_df[var].std()
    median = numeric_df[var].median()
    mode = numeric_df[var].mode()
    max = numeric_df[var].max()
    min = numeric_df[var].min()
    print(f'Mean of {var}: {mean}')
    print(f'Standard deviation of {var}: {std}')
    print(f'Median of {var}: {median}')
    print(f'Mode of {var}: {mode}')
    print(f'Max of {var}: {max}')
    print(f'Min of {var}: {min}')
          
    print('\n')
```

```python

# Convert the "CRASH TIME" column to datetime format
df["CRASH TIME"] = pd.to_datetime(df["CRASH TIME"])

# Extract the hour from the "CRASH TIME" column
df["HOUR"] = df["CRASH TIME"].dt.hour

# Count the occurrences of each hour
hour_counts = df["HOUR"].value_counts()

# Find the top 5 hours with the most collisions
top_5_hours = hour_counts.nlargest(5).reset_index()

# Rename the columns of the top_5_hours DataFrame
top_5_hours.columns = ['Hour', 'Total Collisions']

# Convert the DataFrame to a Markdown table format
markdown_table = top_5_hours.to_markdown(index=False)

# Print the Markdown table
print(markdown_table)
```
